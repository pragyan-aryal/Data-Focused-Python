```python
import pandas as pd
```

```python
code = 'SE'
```

```python
Data1 = code + '.tsv'
Data2 = code + '_desc.tsv'
```

```python
DF1 = pd.read_csv(Data1, sep='\t')
DF1
```

```python
DF2 = pd.read_csv(Data2, sep='\t')
```

```python
columns = ['Key', 'Description', 'Benefits']
```

```python
DF2.columns = columns
DF2
```

```python
Merged = code + '_Merged'
```

```python
Merged = pd.merge(DF1, DF2, on='Key', how='inner')
```

```python
Merged = Merged.drop_duplicates('Key')
```

```python
Merged.to_csv(code + '_final.tsv', sep='\t', index=False)
```
